## Here i detect stuff with nan values, remove features with more than 10% nan and datapoint with less than 10% nvn feautre


```python
train=pd.read_csv(path)
null_values=train.isna().sum()
dropped_features=[]
# print(null_values)
# print(null_values.loc['Id'])
print(null_values.sort_values(ascending=False)/len(train)*100)
print("TOTAL ROWS LENGTHS before IS "+str(len(train)))
for null_indx,null_value in null_values.items():
    percent=(null_value/len(train))
    if  percent<0.1 and percent>0:
        train = train[train[null_indx].notna()]
    elif percent>=0.1:
        train=train.drop(columns=[null_indx])
        dropped_features.append(null_indx)
        
null_values=train.isna().sum()
#         null_values=null_values.drop(labels=[null_indx])
print("TOTAL ROWS LENGTHS IS "+str(len(train)))
print(null_values.sort_values(ascending=False)/len(train)*100)
   
```

mmkay now i wanna see what has high corelation and at the same time have high nan values to see what i wanna do with those

```python
corr_matrix=train.corr()
corr_result=corr_matrix['SalePrice'].sort_values(ascending=False)
corr_result
```

```python
dropped_features
```

Choosing High correlation features

```python
high_corr_features=[]

for indx,result in corr_result.items():
    if(result>0.5):
        high_corr_features.append(indx)
        
high_corr_features
```

making training data

```python
high_corr_features.remove('SalePrice')
x_train,y_train=train[high_corr_features],train['SalePrice']
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(x_train,y_train)
predictions=reg.predict(x_train)
print(predictions[:5])
print(y_train[:5])


plt.plot(y_train[:5], predictions[:5])
```

```python
from sklearn.metrics import mean_absolute_error
val_maee = mean_absolute_error(predictions, y_train,multioutput='raw_values')
print(predictions)
print(val_maee)
print("mean is :",val_maee.mean())
print("std_dev is : ",val_maee.std())
```
